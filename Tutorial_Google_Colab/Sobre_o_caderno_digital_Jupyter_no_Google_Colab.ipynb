## Vantagens

* Aplicação web que roda em um navegador
* Entrega as necessidades de programação e análise de dados
* Não requer download ou instalação, diminuindo a necessidade de suporte
* De fácil uso

## Tipos de células

Os cadernos têm dois tipos principais de células:

* Markdown: textos, imagens, equações, outros recursos
* Código: entrada de códigos que serão executados, em uma linguagem determinada (no nosso caso, trabalharemos com Python)

## Execução de célula de código

1. Colocar código a ser executado
2. Com a célula selecionada, usar `SHIFT + ENTER`

Alternativamente, `CTRL + ENTER` pode ser executado; neste caso, em vez de mover o cursor para a próxima célula, o usuário continuará com a célula atual selecionada.

Na barra de menu existe a opção `Tools` > `Keyboard Shortcuts` que permite ver todos os outros atalhos de teclado para execuções no caderno.
